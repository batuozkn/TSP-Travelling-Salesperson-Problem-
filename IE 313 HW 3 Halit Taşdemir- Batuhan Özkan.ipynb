```python
#Now, lets create a decision variable x_ij which equals 1 ih the salesman visit the city i before j, and 0 ow
x={}
n=15

m = np.zeros((81,81))
choosen_cities = random.sample(range(1,82),15)
print(choosen_cities)
```

```python
for i in range(n):
    for j in range(n):
        if i !=j :
            x[i,j]= pulp.LpVariable(f"x_{i}_{j}",0,1, pulp.LpInteger)
```

```python
# We want to see these cities names and their distances.
filtre = DataFrame["İL NO"].isin(choosen_cities)
df = DataFrame[filtre]
df["Sıralama"] = df["İL NO"].apply(lambda x: choosen_cities.index(x)+1)
df = df.sort_values(by="Sıralama")
df
```

```python
#lets create another decision variable y_k with index k ; and y_k shows the sequence the salesman visit the city k.
y={}
for k in range(1,n):
    y[k]= pulp.LpVariable(f"y_{k}",2,n,pulp.LpInteger)
k=[]
```

```python
for i in range(n):
    for j in range(n):
        if i !=j :
            print(DataFrame.iat[choosen_cities[i],choosen_cities[j]+2])
            
            m[i][j] = DataFrame.iat[choosen_cities[i],choosen_cities[j]+2]
            
```

```python
#Objective function
lp1 += sum(m[i][j]*x[i,j] for i in range(n) for j in range (n) if i!= j)
```

```python
# degree constraints
for i in range(n):
    lp1 += sum(x[i, j] for j in range(n) if i!= j) == 1
for j in range(n):
    lp1 += sum(x[i, j] for i in range(n) if i!= j) == 1
     
```

```python
# subtour elimination constraints 
for i in range(1,n):
    for j in range(1,n):
        if i != j:
            lp1 += (y[i] - y[j] + (n-1) * x[i, j] +  (n-3) * x[j,i]<=n-2)
```

```python
lp1.sense = pulp.LpMinimize
lp1.solve(solver)
print(f"Optimal solution : {pulp.value(lp1.objective)}")
print(f"optimal x values :")
for i in range(n):
    for j in range(n):
        if i != j:
            print(f"x_{i}_{j} = {pulp.value(x[i, j])}")
print(f"optimal y values:")
for k in range (1,n):
    print(f"y_{k} = {pulp.value(y[k])}")
    
```
